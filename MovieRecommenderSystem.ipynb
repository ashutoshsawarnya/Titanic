```python
# First Import libraries


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('white')
```

```python
# then read the dataset

columns=['user_id','item_id','rating','timestamp']
data=pd.read_csv('u.data',sep="\t",names=columns)
```

```python
data.head()
```

```python

# now lets read the movie titles id dataset
movie_titles=pd.read_csv('Movie_Id_Titles')
movie_titles.head()
```

```python
# No we will merge both the dataset on item id
df=pd.merge(data,movie_titles,on="item_id")
df.head()
```

```python
# Now we will EDA part

# Get a look to the best rated movies

# find mean of rating 

df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
# count total number of rating for each movies 

df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
# Make a dataframe of mean rating of movies group by their title name


ratings=pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
# Add num of rating to the dataframe of movies group by title

ratings['num_of_ratings']=pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
# Histrogram for num of ratings of movies

plt.figure(figsize=(10,4))
# ratings['num_of_ratings'].hist(bins=70)
```

```python
# Histrogram for mean rating for each movies


plt.figure(figsize=(10,4))
ratings['rating'].hist(bins=70)
```

```python
# jointplot containing num of ratings and mean ratngs using seaborn

sns.jointplot(x='rating',y='num_of_ratings',data=ratings,alpha=0.5)
```

```python
# Recommending similar movies

# lets make a pivot table for all movies ratings given by user

movie=df.pivot_table(index='user_id',columns='title',values='rating')
movie.head()
```

```python
#most rated movies

ratings.sort_values('num_of_ratings',ascending=False).head(10)
```

```python
# lets select a movie and according to which we recommend movies

ratings_star_wars=movie['Star Wars (1977)']
ratings_star_wars.head()
```

```python
#corelate pivot table to particular movie

similar_starwars=movie.corrwith(ratings_star_wars)
```

```python
# Make a dataframe name with column corelation and fill corelate values in na columns

corr_starwars=pd.DataFrame(similar_starwars,columns=['Corelation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Corelation',ascending=False).head(10)
```

```python
corr_starwars=corr_starwars.join(ratings['num_of_ratings'])


corr_starwars.head()
```

```python
# join num of ratings also to this dataframe and sort corelation values by descending order 
# so we get recommend movies for these type of movies



corr_starwars[corr_starwars['num_of_ratings']>100].sort_values('Corelation',ascending=False).head(10)
```

```python
# Now for second movies

ratings_liar_liar=movie['Liar Liar (1997)']
similar_liar=movie.corrwith(ratings_liar_liar)

corr_liar=pd.DataFrame(similar_liar,columns=['Corelation'])
corr_liar.dropna(inplace=True)
corr_liar=corr_liar.join(ratings['num_of_ratings'])
corr_liar[corr_liar['num_of_ratings']>100].sort_values('Corelation',ascending=False).head(10)
```
